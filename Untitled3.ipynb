```python
import pandas as pd
from datasets import Dataset
from transformers import T5Tokenizer

# Load your CSV dataset
df = pd.read_csv("articles_summaries_updated (18).csv")

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(df)

# Initialize tokenizer
tokenizer = T5Tokenizer.from_pretrained('t5-small')

# Preprocessing function
def preprocess_function(examples):
    inputs = tokenizer(examples['Article'], max_length=512, truncation=True, padding='max_length')
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(examples['Summary'], max_length=150, truncation=True, padding='max_length')

    labels_ids = labels['input_ids']
    # Replace padding token id's with -100 to ignore in loss
    labels_ids = [[(token if token != tokenizer.pad_token_id else -100) for token in label] for label in labels_ids]

    inputs['labels'] = labels_ids
    return inputs

# Tokenize dataset
tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
split_dataset = tokenized_dataset.train_test_split(test_size=0.1)
train_dataset = split_dataset['train']
eval_dataset = split_dataset['test']
```

```python
from transformers import T5ForConditionalGeneration, Trainer, TrainingArguments

model = T5ForConditionalGeneration.from_pretrained('t5-small')
```

```python
training_args = TrainingArguments(
    output_dir='./results',
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    save_total_limit=2,
    save_steps=500,
    # no evaluation_strategy param here
)
```

```python
model = T5ForConditionalGeneration.from_pretrained('t5-small')
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset
)
```

```python
trainer.train()
```

```python
eval_results = trainer.evaluate(eval_dataset)
print(eval_results)
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Make sure model is on the correct device
model = model.to(device)

def generate_summary(text):
    inputs = tokenizer(text, return_tensors="pt", max_length=512, truncation=True).to(device)
    outputs = model.generate(
        **inputs,
        max_length=150,
        num_beams=4,
        early_stopping=True,
    )
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return summary

# Example usage:
new_article = (
    "Artificial Intelligence (AI) has been rapidly evolving over the past decade, impacting various industries including "
    "healthcare, finance, and transportation. Machine learning algorithms, a subset of AI, enable computers to learn from data "
    "and improve their performance without explicit programming. Recent advancements in deep learning have further accelerated "
    "AI capabilities, allowing for more accurate image recognition, natural language processing, and autonomous driving. "
    "Despite the many benefits, AI also poses challenges such as ethical concerns, job displacement, and data privacy issues. "
    "Researchers and policymakers are actively working to address these concerns to ensure responsible AI development."
)

print(generate_summary(new_article))
```

```python
import pytesseract
from PIL import Image

# Open the uploaded image
image_path = "082.png"  # Replace with the uploaded file name

img = Image.open(image_path)

# Use pytesseract to extract text
extracted_text = pytesseract.image_to_string(img)

print("Extracted Text:")
print(extracted_text)
```

```python
import re

# Remove extra spaces, newlines, and special characters
cleaned_text = re.sub(r'\s+', ' ', extracted_text)  # Replace multiple spaces/newlines with a single space
cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)  # Remove special characters

# Print the cleaned text in a single line
print("Cleaned Text:")
print(cleaned_text)
```

```python
import torch

# Automatically select GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Move model to device
model = model.to(device)

# Tokenize the text and move inputs to device
inputs = tokenizer(cleaned_text, return_tensors="pt", max_length=1024, truncation=True)
input_ids = inputs['input_ids'].to(device)

# Generate the summary
summary_ids = model.generate(
    input_ids,
    max_length=300,
    num_beams=4,
    length_penalty=2.0,
    early_stopping=True
)

# Decode and print the summary
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print("Summary:")
print(summary)
```

```python
!pip install transformers --quiet

from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

# Load the pretrained T5 tokenizer and model
tokenizer = T5Tokenizer.from_pretrained("t5-small")
model = T5ForConditionalGeneration.from_pretrained("t5-small")

# Use GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = model.to(device)

def summarize_text(text):
    # T5 expects a prefix like "summarize: "
    input_text = "summarize: " + text.strip()

    # Tokenize and move to device
    inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)
    input_ids = inputs["input_ids"].to(device)

    # Generate summary
    summary_ids = model.generate(
        input_ids,
        max_length=150,
        num_beams=4,
        early_stopping=True
    )

    # Decode and return summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

print("Summary:\n", summarize_text(cleaned_text))
```

```python
from IPython.display import Markdown, display

def print_full_summary(summary):
    display(Markdown(f"**Summary:**\n\n{summary}"))

# Example usage
summary = summarize_text(cleaned_text)
print_full_summary(summary)
```
